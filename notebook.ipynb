# Kidney Dataset
### Data loading and preprocessing

Due to OneHotEncoder we chose to separate beforehand the target from the features

```python
#load kidney dataset
kidney_data = load_kidney_disease_data()

#clean kidney dataset
kidney_data=clean_data(kidney_data)

# load kidney training data and labels
kidney_features_data, kidney_label_data = features_label_data_kidney()

#prepare the data
kidney_features_data = preprocessing_pipeline().fit_transform(kidney_features_data)
kidney_label_data = preprocessing_labels_pipeline().fit_transform(kidney_label_data).ravel()

#split the data
kidney_training, kidney_test, kidney_training_label, kidney_test_label = prepare_dataset(kidney_features_data,kidney_label_data, test_size=0.3)
```

```python
pd.DataFrame(kidney_features_data)
```

```python
plot_features(kidney_data)
```

We see that some features could be very helpful for the classification such as **al, htn and  appet**.

```python
plot_class_repartitions(kidney_data)
```

There are more ckd data so we should be careful in the data repartition, especially in cross validation

### Training and cross validation

Let's run the `train_and_validate_models()` function on this dataset to train different models (SVM, Random Forest, Logistic Regression) and compare the results

```python
kidney_results = train_and_validate_models(kidney_training, kidney_test, kidney_training_label, kidney_test_label)
```

```python
kidney_results
```

```python
visualize_results(kidney_results)
```

We obtained very high results, making it difficult to compare the models directly or improve them further. Among the models with the highest scores, SVM and AdaBoost stand out. However, we believe the SVM should be preferred due to its superior explainability, as it is easier to understand.
